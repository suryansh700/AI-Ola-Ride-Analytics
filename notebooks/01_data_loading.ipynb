```python
missing=df.isnull().sum()

missing
```

```python
missing_percent=(df.isnull().sum()/len(df))*100

missing_report=pd.DataFrame({

"Missing Values":missing,

"Percentage":round(missing_percent,2)

})

missing_report.sort_values(

"Percentage",

ascending=False
)
```

```python
duplicates=df.duplicated().sum()

print("Duplicate Rows :",duplicates)
```

```python
datatype_report=pd.DataFrame({

"Column":df.columns,

"Data Type":df.dtypes.values

})

datatype_report
```

```python
unique_values=pd.DataFrame({

"Column":df.columns,

"Unique Values":df.nunique().values

})

unique_values
```

```python
memory=df.memory_usage(deep=True).sum()/1024**2

print(f"Memory Usage : {memory:.2f} MB")
```

```python
df["Booking Status"].value_counts()
```

```python
status=df["Booking Status"].value_counts()

plt.figure(figsize=(8,5))

status.plot(kind="bar")

plt.title("Booking Status Distribution")

plt.xlabel("Booking Status")

plt.ylabel("Count")

plt.tight_layout()

plt.show()
```

```python
vehicle=df["Vehicle Type"].value_counts()

vehicle
```

```python
plt.figure(figsize=(8,5))

vehicle.plot(kind="bar")

plt.title("Vehicle Type Distribution")

plt.xlabel("Vehicle Type")

plt.ylabel("Count")

plt.tight_layout()

plt.show()
```

```python
payment=df["Payment Method"].value_counts()

payment
```

```python
plt.figure(figsize=(8,5))

payment.plot(kind="pie",autopct="%1.1f%%")

plt.ylabel("")

plt.tight_layout()

plt.show()
```

```python
quality_report=pd.DataFrame({

"Column":df.columns,

"Data Type":df.dtypes.values,

"Missing Values":df.isnull().sum().values,

"Unique Values":df.nunique().values

})

quality_report.to_csv(

"../reports/initial_data_quality_report.csv",

index=False

)

print("Data Quality Report Saved Successfully")
```
