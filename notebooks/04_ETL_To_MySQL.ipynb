```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd

import mysql.connector

from sqlalchemy import create_engine

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("../data/cleaned/ola_cleaned.csv")

print(df.shape)

df.head()
```

```python
HOST = "127.0.0.1"

USER = "root"

PASSWORD = "Surya@123"

DATABASE = "ola_analytics"
```

```python
connection = mysql.connector.connect(

host=HOST,

user=USER,

password=PASSWORD

)

cursor = connection.cursor()

print("Connected Successfully")
```

```python
cursor.execute(

f"""

CREATE DATABASE IF NOT EXISTS {DATABASE};

"""

)

print("Database Created")
```

```python
cursor.execute(

f"USE {DATABASE}"

)

print("Database Selected")
```

```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

connection_url = URL.create(
    drivername="mysql+mysqlconnector",
    username=USER,
    password=PASSWORD,
    host=HOST,
    database=DATABASE
)

engine = create_engine(connection_url)

print("Engine Created Successfully")
```

```python
df.to_sql(

"ola_rides",

engine,

if_exists="replace",

index=False

)

print("Data Imported Successfully")
```

```python
query="""

SELECT COUNT(*)

FROM ola_rides;

"""

count=pd.read_sql(query,engine)

count
```

```python
query="""

SELECT *

FROM ola_rides

LIMIT 10;

"""

sample=pd.read_sql(query,engine)

sample
```

```python
query="""

SHOW COLUMNS

FROM ola_rides;

"""

columns=pd.read_sql(query,engine)

columns
```

```python
query="""

SELECT

COUNT(DISTINCT `Customer ID`)

AS Customers

FROM ola_rides;

"""

pd.read_sql(query,engine)
```

```python
query="""

SELECT

COUNT(DISTINCT `Driver ID`)

AS Drivers

FROM ola_rides;

"""

pd.read_sql(query,engine)
```

```python
query="""

SELECT

SUM(`Booking Value`)

AS Revenue

FROM ola_rides;

"""

pd.read_sql(query,engine)
```

```python
query="""

SELECT

AVG(`Booking Value`)

AS AverageFare

FROM ola_rides;

"""

pd.read_sql(query,engine)
```

```python
query="""

SELECT

AVG(`Ride Distance (km)`)

AS AverageDistance

FROM ola_rides;

"""

pd.read_sql(query,engine)
```

```python
query="""

SELECT

COUNT(*)

FROM ola_rides

WHERE

`Booking Status`='Success';

"""

pd.read_sql(query,engine)
```
